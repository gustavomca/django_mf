```python
sys.path.append(current_dir)
os.environ.setdefault("DJANGO_SETTINGS_MODULE", 'members_frequency.settings')
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
django.setup()
```

```python
from members.models import Pessoa, Membro
```

```python
file_name = 'membros.csv'
file_path = f'{current_dir}/importacoes/{file_name}'

df = pd.read_csv(file_path, sep=';')
df['data_nascimento'] = pd.to_datetime(df['data_nascimento'], format='%d/%m/%Y')
```

```python
df.replace(np.nan, None, inplace=True)
df
```

```python
for _,pessoa in df.iterrows():
    pessoa_igreja = Pessoa.objects.create(
        nome = pessoa['nome'],
        data_nascimento = pessoa['data_nascimento'],
        telefone_residencial = pessoa['telefone_residencial'],
        telefone_celular = pessoa['telefone_celular'],
        endereco_logradouro = pessoa['endereco_logradouro'],
        endereco_numero = pessoa['endereco_numero'],
        endereco_complemento = pessoa['endereco_complemento'],
        endereco_bairro = pessoa['endereco_bairro'],
        endereco_cep = pessoa['endereco_cep'],
        sexo = pessoa['sexo']
    )
    membro = Membro.objects.create(
        membro = pessoa_igreja,
        livro = pessoa['livro'],
        ata = pessoa['ata'],
        folha = pessoa['folha'],
        numero = pessoa['numero'],
        ano = pessoa['ano'],
        data_recepcao = None,
        ativo = True,
        ausente = True if pessoa['ausente']=='x' else False,
        baixa_frequencia = True if pessoa['baixa_frequencia']=='x' else False,
        necessidades_especiais = True if pessoa['necessidades_especiais']=='x' else False,
        profissao_fe = True)
```
